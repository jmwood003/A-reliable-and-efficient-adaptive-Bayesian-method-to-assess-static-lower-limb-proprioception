The degree of agreement (or concordance) between two readings (from the same sample) can be measured by the variation from the 45 degree line through the origin (through the concordance line).

```python
def CCCsim(M1=0,M2=5,C=0.5):
    
    D = np.random.multivariate_normal([M1,M2], [[1,C],[C,1]], size=1000)
    D1 = D[:,0]
    D2 = D[:,1]

    val = np.max(abs(D))
    U = np.linspace(-val,val)

    #calculate the CCC
    var1 = np.var(D1)
    var2 = np.var(D2)
    mu1 = np.mean(D1)
    mu2 = np.mean(D2)
    rho, _ = pearsonr(D1,D2)

    CCC = (2*rho*var1*var2) / (var1+var2 + (mu1-mu2)**2)
    
    print('num = ', str(2*rho*var1*var2))
    print('denom = ', str((var1+var2 + (mu1-mu2)**2)))
    print('CCC = ', str(round(CCC,2)))
    print('Pearsons rho = ', str(round(rho,2)))

    plt.figure()
    plt.plot(D1,D2,'.')
    plt.plot(U,U,'k')
    plt.xlim(-val,val)
    plt.ylim(-val,val)
    plt.show()
    
_ = widgets.interact(CCCsim, M1=(-50, 50, 5),M2=(-50, 50, 5),C=(-1,1,0.1))   
```
